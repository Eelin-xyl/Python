I also did some coding work. For example, for this process, assume that there are three color characteristics: red, yellow, and blue. In machine learning algorithms, we want the machine to distinguish between them, rather than comparing them. Therefore, we will use one-hot encoding. In our data function, "artist names" are independent of each other and have no effect, so we use a single encoder to process them. Finally, they are splitted into training set and test set.

```python
dataset = pd.read_csv('SpotifyAudioFeaturesApril2019.csv')

dataset=dataset.drop('track_name',axis=1)
dataset=dataset.drop('track_id',axis=1)

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

#OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([('artist_name', OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Methods

My study will use SVR, KNN and Linear Regression(from sklearn lib) to predict the target. The mean-absolute-error will be used as the measure of quality of each machine learning model on my problem.

## SVR
SVM is a commonly used classification model whose purpose is to find a hyperplane and predict classification by mapping it to a high-dimensional space. In SVR, we define an ε, as shown in the figure above, the residual error of the data points defining the area within the dotted line is 0, and the distance from the data point (support vector) outside the dotted area to the boundary of the dotted line is the residual (ζ) . Similar to linear models, we want these residuals (ζ) to be minimal. So roughly speaking, SVR is to find an optimal strip area (2ε width), and then return to points outside the area.


```python
# Fitting SVM Regression to the Training set
from sklearn import svm

model_svr = svm.SVR()
model_svr.fit(X_train, y_train)

# Predicting the Test set results
from sklearn.metrics import mean_absolute_error

y_pred_svr = model_svr.predict(X_test)
print(mean_absolute_error(y_test, y_pred_svr))
```

## KNN
The main implementation process of KNN[3]: 1. Calculate the distance between each sample point in the training sample and the test sample (common distance metrics include Euclidean distance, Mahalanobis distance, etc.) 2. Sort all the above distance values 3. Select the first k samples with the smallest distance 4. Vote based on the labels of these k samples to get the final classification category.
Briefly introduce the KNN regression model. There are a series of sample coordinates (x, y), and then given a test point coordinate x1, find the corresponding y1 value on the regression curve. The simplest way is to take the k nearest sample coordinates to x1, and then average their y1 values.

As shown in the figure above, a new x value (red dot in the figure above), k=5 represents the five points closest to x (linked by the red line), and the average of these five points is used as the predicted value of y.

```python
# Fitting KNN Regression to the Training set
from sklearn.neighbors import KNeighborsRegressor

model_knn = KNeighborsRegressor()
model_knn.fit(X_train, y_train)

# Predicting the Test set results
from sklearn.metrics import mean_absolute_error

y_pred_knn = model_knn.predict(X_test)
print(mean_absolute_error(y_test, y_pred_knn))
```

## Linear Regression
The linear regression model tries to learn a linear model to predict the output label Y of the real value X as accurately as possible. In this model, the dependent variable Y is continuous, and the independent variable X can be continuous or discrete. In regression analysis, if only one independent variable and one dependent variable are included, and the relationship between the two can be approximated by a straight line, it is called unary linear regression analysis; if the regression analysis includes two or more independent variables, and the Variables and independent variables have a linear relationship, which is called multiple linear regression analysis. For two-dimensional space linearity is a straight line, for three-dimensional space linearity is a plane, and for multi-dimensional space linearity is a hyperplane.

```python
# Fitting Linear Regression to the Training Set
from sklearn.linear_model import LinearRegression

model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

# Predicting the Test set results
from sklearn.metrics import mean_absolute_error

y_pred_linear = model_linear.predict(X_test)
print(mean_absolute_error(y_test, y_pred_linear))
```

# Experiments and Results

One the other hand, I adjusted the parameters in each model, and used the Grid-Search-CV library. It is actually a greedy algorithm. Giving some parameter option lists. First adjust the parameters that have the greatest impact on the model; Then adjust the next parameter with the greatest impact. Go back and Continue these operations.

```python
#KNN GridSearchCV

#Setting parameter list
from sklearn.model_selection import GridSearchCV

param_grid = [
    {
        'weights' : ['uniform'],
        'n_neighbors' : [i for i in range(1, 100, 20)]
    },
    {
        'weights' : ['distance'],
        'n_neighbors' : [j for j in range(1, 100, 20)],
        'p' : [n for n in range(1, 20, 2)]
    }
]

# Searching for the best parameters
from sklearn.neighbors import KNeighborsRegressor

model_knn = KNeighborsRegressor()
grid_search = GridSearchCV(model_knn, param_grid, n_jobs = -1, verbose = 1)
grid_search.fit(X, y)
print(grid_search.best_params_)
```

The best combination of parameters is [n_neighbors = 10, p = 1, weights = 'distance'].<br/>
Then fit the Training set again.

```python
# Fitting KNN Regression to the Training set
from sklearn.neighbors import KNeighborsRegressor

model_knn_grid = KNeighborsRegressor(n_neighbors = 10, p = 1, weights = 'distance')
model_knn.fit(X_train, y_train)

# Predicting the Test set results
from sklearn.metrics import mean_absolute_error

y_pred_knn_grid = model_knn.predict(X_test)
print(mean_absolute_error(y_test, y_pred_knn_grid))
```
